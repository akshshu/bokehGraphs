```python
from bokeh.io import output_file, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, ranges, LabelSet
# output_file("bar_sorted.html")
from bokeh.palettes import PuBu

# model = ['bagnet9', 'bagnet17', 'bagnet33', 'Vgg16', 'Vgg19', 'Resnet50']
# acc = [67.12, 79.97, 88.16, 94.51, 94.84, 96.59]
source = ColumnDataSource(dict(x=['bagnet9', 'bagnet17', 'bagnet33', 'Vgg16', 'Vgg19', 'Resnet50'],y=[41.18, 56.36, 67.67, 79.42, 79.7, 83.67]))
x_label = "Models"
y_label = 'Top-1 Accuracy(%)'
title = "Impact of Receptive field size on Top-1 Accuracy of different Models"
# sorting the bars means sorting the range factors
# sorted_acc = sorted(model, key=lambda x: acc[model.index(x)])
p = figure(plot_width=700, plot_height=550, tools="save",
        x_axis_label = x_label,
        y_axis_label = y_label,
        title=title,
        x_minor_ticks=2,
        x_range = source.data["x"],
        y_range= ranges.Range1d(start=0,end=100))
labels = LabelSet(x='x', y='y', text='y', level='glyph',
        x_offset=-15.5, y_offset=0, source=source, render_mode='canvas',text_font_size='8pt',text_font_style='bold')

p.vbar(source=source,x='x',top='y',bottom=0,width=0.3,color="navy")
p.add_layout(labels)
p.title.text_font_size = '9pt'
p.xaxis.axis_label_text_font_size = "9pt"
p.xaxis.axis_label_text_font_style= "bold"
p.yaxis.axis_label_text_font_size = "9pt"
p.yaxis.axis_label_text_font_style= "bold"
p.axis.axis_label_text_font_style = 'bold'
p.axis.major_label_text_font_style = 'bold'

# p.xgrid.grid_line_color = None
# p.y_range.start = 0

# p.vbar(source=source, x='x', width=0.3)

show(p)

# bokeh.plotting.show(p)
```

```python

from bokeh.models import NumeralTickFormatter
from bokeh.plotting import figure, show

# prepare some data
p_size= [1, 2, 3, 5,10,15]
clean_acc = [95.96,95.96,96.06,95.91,95.79,95.66]
# x2 = [1, 2, 3, 5,10,15]
clean_acc2 = [75.68,74.96,74.11,72.20,70.65,67.7]

source = ColumnDataSource(data=dict(x_pt=p_size,
                                    y_pt=clean_acc,
                                    y_val=clean_acc))
source2 = ColumnDataSource(data=dict(x_pt2=p_size,
                                    y_pt2=clean_acc2,
                                    y_val2=clean_acc2))
# create new plot
p = figure(
    title="Top5 Clean Accuracy comparison on Original and Modified dataset on different Adversarial Patch sizes",
    sizing_mode="stretch_width",
    max_width=800,
    height=300,
    y_range= ranges.Range1d(start=0,end=100)
)
labels = LabelSet(x='x_pt', y='y_pt', text='y_pt',
              x_offset=5, y_offset=-20, source=source, render_mode='canvas',text_font_size='8pt',text_font_style='bold')
labels2 = LabelSet(x='x_pt2', y='y_pt2', text='y_pt2',
              x_offset=5, y_offset=-20, source=source2, render_mode='canvas',text_font_size='8pt',text_font_style='bold')
p.title.text_font_size = '9pt'
# format axes ticks
p.yaxis[0].formatter = NumeralTickFormatter()

# add renderers


p.circle(p_size, clean_acc, size=8,color="navy")
p.line(p_size, clean_acc, color="navy", line_width=1, legend_label="Top5 Clean Accuracy on Orig. Dataset")
p.circle(p_size, clean_acc2, size=8,color="red")
p.line(p_size, clean_acc2, color="red", line_width=1, legend_label="Top5 Clean Accuracy on Mod. Dataset")
p.legend.label_text_font_size='8pt'
p.legend.label_text_font_style = "bold"
p.xaxis.axis_label = 'Adversarial Patch Size (% of Image Size)'
p.yaxis.axis_label = 'Top1 Clean Accuracy (in %)'
p.legend.location = "center"

p.xaxis.axis_label_text_font_size = "8pt"
p.xaxis.axis_label_text_font_style= "bold"
p.yaxis.axis_label_text_font_size = "8pt"
p.yaxis.axis_label_text_font_style= "bold"
p.axis.major_label_text_font_style = 'bold'

# show the results

p.add_layout(labels)
p.add_layout(labels2)
show(p)
```

```python
from bokeh.models import NumeralTickFormatter
from bokeh.plotting import figure, show

# prepare some data
p_size = [1, 2, 3, 5,10,15]
prov_acc = [74.01,69.47,63.42,47.07,26.33,8.71]
# p_size2 = [1, 2, 3, 5,10,15]
prov_acc2 = [44.16,39.48,34.55,22.20,11.79,4.63]
source = ColumnDataSource(data=dict(x_pt=p_size,
                                    y_pt=prov_acc,
                                    y_val=prov_acc))
source2 = ColumnDataSource(data=dict(x_pt2=p_size,
                                    y_pt2=prov_acc2,
                                    y_val2=prov_acc2))
# create new plot
p = figure(
    title="Top5 Provable Robust Accuracy comparison on Original and Modified dataset on different Adversarial Patch sizes",
    sizing_mode="stretch_width",
    max_width=800,
    height=300,
    y_range= ranges.Range1d(start=0,end=100)
)
labels = LabelSet(x='x_pt', y='y_pt', text='y_pt',
              x_offset=5, y_offset=2, source=source, render_mode='canvas',text_font_size='8pt',text_font_style='bold')
labels2 = LabelSet(x='x_pt2', y='y_pt2', text='y_pt2',
              x_offset=5, y_offset=2, source=source2, render_mode='canvas',text_font_size='8pt',text_font_style='bold')
p.title.text_font_size = '9pt'
# format axes ticks
p.yaxis[0].formatter = NumeralTickFormatter()

# add renderers

p.circle(p_size, prov_acc, size=8,color="navy")
p.line(p_size, prov_acc, color="navy", line_width=1, legend_label="Top5 Provable Robust Accuracy on Orig. Dataset")
p.circle(p_size, prov_acc2, size=8,color="red")
p.line(p_size, prov_acc2, color="red", line_width=1, legend_label="Top5 Provable Robust Accuracy on Mod. Dataset")
p.legend.label_text_font_size='8pt'
p.legend.label_text_font_style = "bold"
p.xaxis.axis_label = 'Adversarial Patch Size (% of Image Size)'
p.yaxis.axis_label = 'Top5 Provable Robust Accuracy (in %)'

p.xaxis.axis_label_text_font_size = "8pt"
p.xaxis.axis_label_text_font_style= "bold"
p.yaxis.axis_label_text_font_size = "8pt"
p.yaxis.axis_label_text_font_style= "bold"
p.axis.major_label_text_font_style = 'bold'

# show the results

p.add_layout(labels)
p.add_layout(labels2)
show(p)
```

```python
from bokeh.models import ColumnDataSource, Label, LabelSet, Range1d
from bokeh.plotting import figure, output_file, show

output_file("label.html", title="label.py example")

source = ColumnDataSource(data=dict(height=[66, 71, 72, 68, 58, 62],
                                    weight=[165, 189, 220, 141, 260, 174],
                                    names=['Mark', 'Amir', 'Matt', 'Greg',
                                           'Owen', 'Juan']))

p = figure(title='Dist. of 10th Grade Students at Lee High',
           x_range=Range1d(140, 275))
p.scatter(x='weight', y='height', size=8, source=source)
p.xaxis[0].axis_label = 'Weight (lbs)'
p.yaxis[0].axis_label = 'Height (in)'

labels = LabelSet(x='weight', y='height', text='names',
              x_offset=5, y_offset=5, source=source, render_mode='canvas')

citation = Label(x=70, y=70, x_units='screen', y_units='screen',
                 text='Collected by Luke C. 2016-04-01', render_mode='css',
                 border_line_color='black', border_line_alpha=1.0,
                 background_fill_color='white', background_fill_alpha=1.0)

p.add_layout(labels)
p.add_layout(citation)

show(p)
```
